```python
import pandas as pd
```

```python
# 칼럼명 지정
column_names = [
    'date', 'card_type', 'card_number', 'seq', 'in_out', 
    'transport_id', 'transport_name', 'station_id', 'station_name',
    'device_id', 'transaction_id', 'payment_type', 'time',
    'passenger_type', 'passenger_detail', 'amount', 'extra_info'
]

df = pd.read_csv(
    r"C:\Users\HP\Desktop\작업 폴더\01.work\리버풀\데이터\A001_202507\A001_202507.csv",
    names=column_names,
    header=0,
    dtype=str,
    low_memory=False
)

# dtype 최적화
# 날짜/시간
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d', errors='coerce')
df['time'] = pd.to_numeric(df['time'], errors='coerce').astype('int16')  # HHMM 형식

# 계산용 숫자
df['seq'] = pd.to_numeric(df['seq'], errors='coerce').astype('int16')
df['amount'] = pd.to_numeric(df['amount'], errors='coerce').astype('float32')

# 식별자 - category
df['card_type'] = df['card_type'].astype('category')
df['card_number'] = df['card_number'].astype('category')
df['in_out'] = df['in_out'].astype('category')
df['transport_id'] = df['transport_id'].astype('category')
df['transport_name'] = df['transport_name'].astype('category')
df['station_id'] = df['station_id'].astype('category')
df['station_name'] = df['station_name'].astype('category')
df['device_id'] = df['device_id'].astype('category')
df['transaction_id'] = df['transaction_id'].astype('category')
df['payment_type'] = df['payment_type'].astype('category')
df['passenger_type'] = df['passenger_type'].astype('category')
df['passenger_detail'] = df['passenger_detail'].astype('category')
df['extra_info'] = df['extra_info'].astype('category')

print("dtype 변환 완료")
```

```python
# 메모리 사용량 확인
print(f"메모리 사용량: {df.memory_usage(deep=True).sum() / 1024**3:.2f} GB")
print(f"데이터 shape: {df.shape}")
print(f"\n칼럼별 메모리:")
for col in df.columns:
    mem = df[col].memory_usage(deep=True) / 1024**2
    print(f"{col:20s}: {mem:8.2f} MB ({df[col].dtype})")
```

```python
# 커널 재시작 후 이 셀 실행
# Parquet으로 저장 (압축 + 빠른 로딩)
output_parquet = r"C:\Users\HP\Desktop\작업 폴더\01.work\리버풀\데이터\A001_202507\A001_202507_processed.parquet"

df.to_parquet(output_parquet, index=False, compression='snappy')
print(f"Parquet 저장 완료: {output_parquet}")

import os
if os.path.exists(output_parquet):
    size_gb = os.path.getsize(output_parquet) / 1024**3
    print(f"파일 크기: {size_gb:.2f} GB")
    print(f"원본 CSV 대비: {17 / size_gb:.1f}배 작음")
```
